```python
df['actual_price'] = pd.Series(df['actual_price'])
```

```python
df['actual_price'] = df['actual_price'].astype(str).fillna('')
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)
```

```python
df['actual_price']
```

```python
df['discounted_price'] = pd.Series(df['discounted_price'])
df['discounted_price'] = df['discounted_price'].astype(str).fillna('')
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
```

```python
df['discounted_price']
```

```python
df['discount_percentage']
```

```python
df['discount_percentage'] = pd.Series(df['discount_percentage'])
df['discount_percentage'] = df['discount_percentage'].astype(str).fillna('')
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').str.replace(',', '').astype(int)
```

```python
df['discount_percentage']
```

```python
df['rating']
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating']
```

```python
df['rating_count']
df['rating_count'] = pd.Series(df['rating_count'])
df['rating_count'] = df['rating_count'].astype(str).fillna('')
df['rating_count'] = df['rating_count'].str.replace(',', '').astype(float)
df['rating_count']
```

```python
df.drop('user_name',axis=1, inplace=True)
df.drop('product_link',axis=1, inplace=True)
df
```

```python
import plotly.express as px
avg_discounted_price = df.groupby('category')['discounted_price'].mean()

fig = px.bar(
    avg_discounted_price,
    x=avg_discounted_price.values,
    y=avg_discounted_price.index,
    title='Average Discounted Price per Category',
    labels={'x': 'Average Discounted Price', 'y': 'Category'},
    orientation='h',
    color=avg_discounted_price.values,
    color_continuous_scale='Viridis'
)

fig.update_layout(
    yaxis={'categoryorder':'total ascending', 'tickfont': {'size': 10}},
    title={'x': 0.5},
)

fig.show()
```

```python
df.plot.scatter(x='discounted_price', y='rating', title='Discounted Price vs. Rating')
```

```python
fig = px.scatter(
    df,
    x='actual_price',
    y='discounted_price',
    title='Actual Price vs. Discounted Price',
    labels={'actual_price': 'Actual Price', 'discounted_price': 'Discounted Price'},
    hover_data=['product_name'],
    color='discounted_price',
    color_continuous_scale='Viridis'
)

# Update layout for better aesthetics and expanded scale
fig.update_layout(
    title={'x': 0.5},
    xaxis_title='Actual Price',
    yaxis_title='Discounted Price',
    plot_bgcolor='rgba(0,0,0,0)',
    xaxis=dict(showgrid=False, range=[df['actual_price'].min() * 0.9, df['actual_price'].max() * 1.1]),
    yaxis=dict(showgrid=False, range=[df['discounted_price'].min() * 0.9, df['discounted_price'].max() * 1.1])
)

# Show the plot
fig.show()
```

```python
correlation_matrix = df[['actual_price', 'rating_count']].corr()

print(correlation_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix for Actual Price and Rating Count')
plt.show()
```

```python
top_selling_products = df.sort_values(by='rating_count', ascending=False).head(10)

top_selling_products = top_selling_products[['product_name', 'rating', 'rating_count']]
top_selling_products.reset_index(drop=True, inplace=True)
top_selling_products

```

```python
plt.figure(figsize=(8, 8))
plt.pie(top_selling_products['rating_count'], labels=top_selling_products['product_name'], autopct='%1.1f%%', startangle=140)
plt.title('Top Selling Products by Rating Count')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
s pricing (both actual and discounted) affects product ratings and popularity?

fig, axes = plt.subplots(1, 2, figsize=(18, 6))

#Scatter plot for actual price vs rating
sns.scatterplot(ax=axes[0], data=df, x='actual_price', y='rating', size='rating_count', alpha=0.5)
axes[0].set_title('Actual Price vs Rating')
axes[0].set_xlabel('Actual Price')
axes[0].set_ylabel('Average Rating')
axes[0].set_xscale('log')
axes[0].grid(True)

#Scatter plot for discounted price vs rating
sns.scatterplot(ax=axes[1], data=df, x='discounted_price', y='rating', size='rating_count', alpha=0.5)
axes[1].set_title('Discounted Price vs Rating')
axes[1].set_xlabel('Discounted Price')
axes[1].set_ylabel('Average Rating')
axes[1].set_xscale('log')  
axes[1].grid(True)

plt.show()
```
